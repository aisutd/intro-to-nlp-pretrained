# Sentiment analysis
from amazon, we copied some comments, and here is a showcase about useing a pretrained NLP model to identify sentiment of comments

image.png

image.png

```
## download pretrained models
classifier = pipeline('sentiment-analysis')
```

```
## here are three positive comments and three negative comments
comment_stars_5 = [
                   r'Wonderful! Comes exactly as pictured in the listing in a plastic case, the cookies also have a plastic insert to stop them from being crushed. Excellent for lunches and take along snacks! I purchased these May 12, 2017 and they expire December 2017, perfect for our needs.',
                   r"You know when you're just having a bad day and you need a little pick me up, but you don't want to be a tubby tub? I love these individually wrapped Milanos because I can commit to two at a time. Yes yes, I sometimes have four...but at the end of the day, as a forever alone kind of guy...these work because they won't go stale.",
                   r"My mom has been eating Dark Chocolate Milano cookies for years. I purchased one package for her house and one for ours! I love that these are individually wrapped with 2 cookies per package. The plastic tub is awesome for keeping them fresh."
                   ]

comment_stars_1 = [
                   r'Oh my word, can we say disappointing? These cookies are individually wrapped, 2 cookies per pack. I was thinking theyd be fresh & delicious as when you buy them in the bag. NO. These are not even Milanos. They are SO dry and full of air they seriously taste awful. Theyre like a dried, flavorless old dehydrated marshmallow. And theres also very very little chocolate. And the cookie is a lot whiter than normal. Id love a refund or replacement. I threw the whole box out after trying a few packs.',
                   r"These used to be a pretty good value until the price went up. At 6.99, it's a bit overpriced for what you get.",
                   r'I would never have purchased these had I know the expiration date would be dec 2018 (its mid October now). Last time I bought these at a bulk store in the summer, the expiration date was 8 months later. Thats more like it! We will never finish these cookies by December...'
                   ]


## show prediction by our model
for comment in comment_stars_5:
  result = classifier(comment)
  print('five star comments are predicted to be:', result)

for comment in comment_stars_1:
  result = classifier(comment)
  print('one star comments are predicted to be:', result)
  
```

#Summerization
## -- extracting the main contents

```
## download pretrained summarization models
summarizer = pipeline("summarization")
```

```
content = r"Reading others reviews, I could tell some buyers were not familiar with Milano Cookies. Some complaints were that there wasnt enough chocolate... \
that is the way they make them, on purpose! Some complained that the use by date was only 3 months. That is standard, that is NOT stale! Some reviews thought the cookies were not crisp enough. \
Milano is a thin short bread cookie. Each bag contains 3 layers with 4 cookies on each layer. That is 12 cookies per bag OR a total of 36 cookies for $8 plus and free shipping. \
My delivery was free next day delivery. Glad to say only about 10-12 % of the cookies had broken in half. No chocolate melted outside of the cookie. \
My Milano bags shipped from Ft Myers, Florida to Port Charlotte, Florida in 97 weather that felt like 110 or more! \
Maybe the condition becomes more melted and broken the further they travel! Am enjoying my Milano cookies so much!!"
result = summarizer(content,min_length=50, max_length=200)
print(result)
```

# Text generation
give some context to the generator, it will generate the rest of that sentence

```
generator = pipeline("text-generation")
```

```
your_context = 'I am super man '
for i in range(1,4):
  result = generator(your_context)
  print()
  print(f"sentence{i}:")
  print(result[0]['generated_text'])
```
